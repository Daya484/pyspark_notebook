UNION and UNION BY NAME


Preaparing Dataframes

```python
data1 = [('1','kad'),
        ('2','sid')]
schema1 = 'id STRING, name STRING' 

df1 = spark.createDataFrame(data1,schema1)

data2 = [('3','rahul'),
        ('4','jas')]
schema2 = 'id STRING, name STRING' 

df2 = spark.createDataFrame(data2,schema2)
```

```python
df1.display()
df2.display()
```

Union

```python
df1.union(df2).display()
```

```python
data1 = [('kad','1',),
        ('sid','2',)]
schema1 = 'name STRING, id STRING' 

df1 = spark.createDataFrame(data1,schema1)

df1.display()
```

Union by Name

```python
df1.unionByName(df2).display()
```

String Functions


Initcap()

```python
from pyspark.sql.functions import *
df.select(initcap('Item_Type').alias('upper_Item_Type')).display()
```

UPPER

```python
from pyspark.sql.functions import *
df.select(upper('Item_Type').alias('upper_Item_Type')).display()
```

lower

```python
df.select(lower('Item_Type').alias('upper_Item_Type')).display()
```

Date Functions

Current_Date

Scinerio-1

```python
df = df.withColumn('curr_date',current_date())

df.display()
```

Scinerio-2

```python
df = df.withColumn('curr_timestamp',current_timestamp())

df.display()
```

Date_Add()

```python
df = df.withColumn('week_after',date_add('curr_date',7))
df.display()
```

Date_Sub()

```python
df=df.withColumn('week_before',date_sub('curr_date',7))
df.display()
```

Date_Format()

```python
df = df.withColumn('week_before',date_format('week_before','dd-MM-yyyy'))
df.display()
```

SPLIT and Indexing


SPLIT

```python
df.withColumn('Outlet_Type',split('Outlet_Type',' ')).display()
```


Indexing

```python
df.withColumn('Outlet_Type',split('Outlet_Type',' ')[1]).display()
```


Explode

```python
df_exp = df.withColumn('Outlet_Type',split('Outlet_Type',' '))

df_exp.display()
```

```python
df_exp.withColumn('Outlet_Type',explode('Outlet_Type')).display()
```

```python
df_exp.withColumn('Type1_flag',array_contains('Outlet_Type','Type1')).display()
```
